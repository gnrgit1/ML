```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
def estimate_coef(x, y):
    n = np.size(x)
    m_x = np.mean(x)
    m_y = np.mean(y)
    ss_xy = np.sum(y * x) - n * m_y * m_x
    ss_xx = np.sum(x * x) - n * m_x * m_x
    b_1 = ss_xy / ss_xx
    b_0 = m_y - b_1 * m_x
    return (b_0, b_1)


def plot_regression_line(x, y, b):
    plt.scatter(x, y, color="m", marker="o", s=30)
    y_pred = b[0] + b[1] * x
    plt.plot(x, y_pred, color="g")
    plt.xlabel('year_of_experience')
    plt.ylabel('salary')
    plt.show()

def main():

    df = pd.read_csv("//home//pg//Downloads//Salary_Data.csv")
    x = df.iloc[:, 0].values
    y = df.iloc[:, 1].values
    b = estimate_coef(x, y)
    plot_regression_line(x, y, b)

if __name__ == "__main__":
    main()
```
